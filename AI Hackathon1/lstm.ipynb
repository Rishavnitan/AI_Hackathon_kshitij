```python
# ============================================================
# IMPORTS
# ============================================================
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import matplotlib.pyplot as plt

# ============================================================
# CONFIG
# ============================================================
SEQ_LEN = 30
BATCH_SIZE = 64
EPOCHS = 20
LR = 5e-4
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
MAX_RUL = 125

# ============================================================
# LOAD DATA (FD001 ONLY)
# ============================================================
cols = (
    ["unit", "cycle"] +
    [f"op_{i}" for i in range(1, 4)] +
    [f"s_{i}" for i in range(1, 22)]
)

train_df = pd.read_csv("train_FD001.txt", delim_whitespace=True, header=None, names=cols)
test_df  = pd.read_csv("test_FD001.txt",  delim_whitespace=True, header=None, names=cols)
rul_df   = pd.read_csv("RUL_FD001.txt", header=None, names=["RUL"])

print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)

# ============================================================
# RUL CALCULATION
# ============================================================
max_cycles = train_df.groupby("unit")["cycle"].max()
train_df["RUL"] = train_df.apply(
    lambda row: max_cycles[row["unit"]] - row["cycle"], axis=1
)
train_df["RUL"] = train_df["RUL"].clip(upper=MAX_RUL)

# ============================================================
# FEATURE SCALING
# ============================================================
features = cols[2:]
scaler = MinMaxScaler()

train_df[features] = scaler.fit_transform(train_df[features])
test_df[features]  = scaler.transform(test_df[features])

# ============================================================
# ENGINE-WISE TRAIN / VALIDATION SPLIT
# ============================================================
engine_ids = train_df["unit"].unique()

train_engines, val_engines = train_test_split(
    engine_ids, test_size=0.2, random_state=42
)

train_split_df = train_df[train_df["unit"].isin(train_engines)].reset_index(drop=True)
val_split_df   = train_df[train_df["unit"].isin(val_engines)].reset_index(drop=True)

print(f"Train Engines: {len(train_engines)}")
print(f"Validation Engines: {len(val_engines)}")

# ============================================================
# SEQUENCE GENERATION
# ============================================================
def generate_sequences(df, seq_len, is_test=False):
    X, y = [], []
    for unit in df["unit"].unique():
        unit_df = df[df["unit"] == unit]
        data = unit_df[features].values

        if is_test:
            X.append(data[-seq_len:])
        else:
            rul = unit_df["RUL"].values
            for i in range(len(data) - seq_len):
                X.append(data[i:i + seq_len])
                y.append(rul[i + seq_len - 1])

    return np.array(X), np.array(y)

X_train, y_train = generate_sequences(train_split_df, SEQ_LEN)
X_val, y_val     = generate_sequences(val_split_df, SEQ_LEN)
X_test, _        = generate_sequences(test_df, SEQ_LEN, is_test=True)

print("Train sequences:", X_train.shape)
print("Validation sequences:", X_val.shape)
print("Test sequences:", X_test.shape)

# ============================================================
# DATASET & DATALOADER
# ============================================================
class CMAPSSDataset(Dataset):
    def __init__(self, X, y=None):
        self.X = torch.tensor(X, dtype=torch.float32)
        self.y = None if y is None else torch.tensor(y, dtype=torch.float32)

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        if self.y is None:
            return self.X[idx]
        return self.X[idx], self.y[idx]

train_loader = DataLoader(CMAPSSDataset(X_train, y_train),
                          batch_size=BATCH_SIZE, shuffle=True)

val_loader   = DataLoader(CMAPSSDataset(X_val, y_val),
                          batch_size=BATCH_SIZE, shuffle=False)

test_loader  = DataLoader(CMAPSSDataset(X_test),
                          batch_size=BATCH_SIZE, shuffle=False)

# ============================================================
# LSTM MODEL
# ============================================================
class RULLSTM(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, num_layers=2, dropout=0.2):
        super().__init__()

        self.lstm = nn.LSTM(
            input_size=input_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout
        )

        self.regressor = nn.Sequential(
            nn.Linear(hidden_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )

    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        last_hidden = lstm_out[:, -1, :]
        return self.regressor(last_hidden).squeeze(1)

model = RULLSTM(len(features)).to(DEVICE)
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=LR)

# ============================================================
# TRAINING WITH VALIDATION
# ============================================================
for epoch in range(EPOCHS):
    model.train()
    train_losses = []

    for X, y in tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}"):
        X, y = X.to(DEVICE), y.to(DEVICE)
        optimizer.zero_grad()
        preds = model(X)
        loss = criterion(preds, y)
        loss.backward()
        optimizer.step()
        train_losses.append(loss.item())

    train_mse = np.mean(train_losses)

    model.eval()
    val_preds, val_true = [], []
    with torch.no_grad():
        for X, y in val_loader:
            X, y = X.to(DEVICE), y.to(DEVICE)
            preds = model(X)
            val_preds.append(preds.cpu().numpy())
            val_true.append(y.cpu().numpy())

    val_rmse = np.sqrt(mean_squared_error(
        np.concatenate(val_true), np.concatenate(val_preds)
    ))

    print(f"Epoch {epoch+1} | Train MSE: {train_mse:.4f} | Val RMSE: {val_rmse:.4f}")

# ============================================================
# EVALUATION (TRAIN / VAL / TEST)
# ============================================================
def evaluate_model(model, dataloader):
    model.eval()
    preds, true = [], []

    with torch.no_grad():
        for batch in dataloader:
            if isinstance(batch, (list, tuple)):
                X, y = batch
                y = y.to(DEVICE)
                true.append(y.cpu().numpy())
            else:
                X = batch

            X = X.to(DEVICE)
            outputs = model(X)
            preds.append(outputs.cpu().numpy())

    preds = np.concatenate(preds)
    true  = None if len(true) == 0 else np.concatenate(true)
    return preds, true

train_preds, train_true = evaluate_model(model, train_loader)
val_preds, val_true     = evaluate_model(model, val_loader)
test_preds, _           = evaluate_model(model, test_loader)

train_rmse = np.sqrt(mean_squared_error(train_true, train_preds))
val_rmse   = np.sqrt(mean_squared_error(val_true, val_preds))
test_rmse  = np.sqrt(mean_squared_error(rul_df["RUL"].values, test_preds))

print("\n==============================")
print(f"TRAIN RMSE: {train_rmse:.4f}")
print(f"VAL   RMSE: {val_rmse:.4f}")
print(f"TEST  RMSE: {test_rmse:.4f}")
print("==============================")

# ============================================================
# RESULTS TABLE
# ============================================================
results_df = pd.DataFrame({
    "Engine_ID": np.arange(1, len(test_preds) + 1),
    "Predicted_RUL": test_preds,
    "Health_Percentage": np.clip((test_preds / MAX_RUL) * 100, 0, 100)
})

print(results_df.head())

# ============================================================
# HEALTH DEGRADATION PLOT
# ============================================================
def plot_engine_health(engine_id, train_df, model):
    engine_df = train_df[train_df["unit"] == engine_id].sort_values("cycle")
    health_curve = []

    model.eval()
    with torch.no_grad():
        for i in range(len(engine_df) - SEQ_LEN):
            seq = engine_df.iloc[i:i + SEQ_LEN][features].values
            seq = torch.tensor(seq, dtype=torch.float32).unsqueeze(0).to(DEVICE)
            pred_rul = model(seq).cpu().item()
            health_curve.append(np.clip((pred_rul / MAX_RUL) * 100, 0, 100))

    cycles = engine_df["cycle"].values[SEQ_LEN:]

    plt.figure(figsize=(10, 5))
    plt.plot(cycles, health_curve, linewidth=2)
    plt.xlabel("Cycle")
    plt.ylabel("Health Percentage (%)")
    plt.title(f"Engine #{engine_id} – Health Degradation Curve")
    plt.ylim(0, 100)
    plt.grid(True)
    plt.show()

# ============================================================
# USER INPUT
# ============================================================
engine_id = int(input(f"Enter Engine ID (1 to {len(results_df)}): "))

if engine_id < 1 or engine_id > len(results_df):
    raise ValueError("Invalid Engine ID")

engine_row = results_df.iloc[engine_id - 1]

print("\n==============================")
print(f"ENGINE #{engine_id} STATUS")
print("------------------------------")
print(f"Predicted Remaining Useful Life : {engine_row['Predicted_RUL']:.2f}")
print(f"Health Percentage              : {engine_row['Health_Percentage']:.2f}%")
print("==============================")

plot_engine_health(engine_id, train_df, model)
```
